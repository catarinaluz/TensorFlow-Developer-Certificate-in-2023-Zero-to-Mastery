#Creating random tensors

```python
#Create two random tensors

random_1 = tf.random.Generator.from_seed(42) #set seed for reproducibility
random_1 = random_1.normal(shape = (3,2))
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))

random_1,random_2,random_1 == random_2
```

#Shuffling the elements in a tensor

```python
not_shuffled = tf.constant([[10,7],
                            [3,4],
                            [2,5]])

tf.random.shuffle(not_shuffled)
```

```python
tf.random.set_seed(42) #global level random seed
tf.random.shuffle(not_shuffled, seed = 42) #operation level random seed
```

#Getting information from tensors

```python
#Create a rank 4 tensor

rank_4_tensor = tf.zeros(shape=[2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
print("Datatype of every element",rank_4_tensor.dtype)
print("Rank:", rank_4_tensor.ndim)
print("Shape:", rank_4_tensor.shape)
print("Elements along 0 axis:", rank_4_tensor.shape[0])
print("Elements along last axis:",rank_4_tensor.shape[-1])
print("Total of elements:",tf.size(rank_4_tensor).numpy())
```



#Indexing tensors


```python
#Get the first 2 elements of each dimension

rank_4_tensor[:2,:2,:2,:2]
```

```python
#Get first element from each dimension from each index except for the final one

rank_4_tensor[:1,:1,:1, :]
```

```python
# Create a rank 2 tensor

rank_2_tensor = tf.constant([[10,7],
                             [3,4]])
rank_2_tensor
```

```python
#Alternative to tf.newaxis

tf.expand_dims(rank_2_tensor,axis = -1)
```

```python


tf.expand_dims(rank_2_tensor,axis = 0)
```

```python
#Create a tensor (3,2)

X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])

Y = tf.constant([[7,8],
                 [9,10],
                 [11,12]])
X,Y
```

```python
#Try to multiply
X@Y
```

```python
tf.matmul(X,Y)
```

```python
tf.reshape(Y,shape = (2,3))
```

```python
#Multiply X by reshape Y

X @ tf.reshape(Y,shape=(2,3))
```

```python
#Change the shape of X instead of Y 
tf.matmul(tf.reshape(X,shape = (2,3)), Y)
```

```python
# Can do the same with transpose
X,tf.transpose(X), tf. reshape(X,shape=(2,3))
```

```python
tf.matmul(tf.transpose(X),Y)
```

```python
#Perform the dot product on X and Y

tf.tensordot(tf.transpose(X),Y,axes=1)
```

```python
tf.matmul(X,tf.transpose(Y))
```

```python
tf.matmul(X,tf.reshape(Y,shape=(2,3)))
```

#Changing the datatype of tensors

```python
#Create a new tensor with default datatype float32
B = tf.constant([1.7,7.4])
B.dtype
```

```python
C = tf.constant([7,10])
C.dtype
```

```python
#float32 to float16 (reduce precision)

D = tf.cast(B, dtype=tf.float16)
D
```

```python
#Change from int32 to float32
E = tf.cast(C,dtype=tf.float32)
E
```

#Aggregating tensors

```python
#Create a random tensor with values between 0 and 100 of size 15

E= tf.constant(np.random.randint(0,110,size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
#Find the minimum
tf.reduce_min(E)
```

```python
#Find the maximum
tf.reduce_max(E)
```

```python
#Find the mean
tf.reduce_mean(E)
```

```python
#Find the sum
tf.reduce_sum(E)
```

```python
#Find the variance
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
#Find the standard deviation

tf.math.reduce_std(tf.cast(E,dtype=tf.float32))
```

#Find the positional maximum and minimum

```python
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
#Find the positional maximum
tf.argmax(F)
```

```python
#Index on our largest value position
F[tf.argmax(F)]
```

```python
#Find the max value of F
tf.reduce_max(F)
```

```python
#Find the positional minimum
tf.argmin(F)
```

```python
#Find the minimum using the positional minimum index
F[tf.argmin(F)]
```

#One-hot encoding tensors

```python
#Create a list of indices
some_list = [0,1,2,3] #could be red,green,blue,purple

#one hot encode our list
tf.one_hot(some_list, depth=4)
```

```python
#Specify custom values for one hot encoding
tf.one_hot(some_list,depth=4, on_value = 'ON', off_value = 'OFF')
```
