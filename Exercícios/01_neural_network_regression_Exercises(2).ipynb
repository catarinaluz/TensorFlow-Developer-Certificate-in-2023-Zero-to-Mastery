```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# Import the Boston Housing dataset

(X_train , y_train) , (X_test , y_test) = tf.keras.datasets.boston_housing.load_data(path = 'boston_housing_npz' , 
                                                                           test_split = 0.3 , seed = 42)
     
```

**First experiment:** 2 layers, the first one with 100 neurons, and 100 epochs

```python

tf.random.set_seed(42)

model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1)
])

model1.compile(loss = tf.keras.losses.mae,
                        optimizer = tf.keras.optimizers.Adam(), 
                        metrics = ["mae"])

history1 = model1.fit(X_train,y_train,epochs = 100, verbose = 0)
```

```python
ev1 = model1.evaluate(X_test,y_test)
  
```

```python
pd.DataFrame(history1.history).plot()
plt.ylabel('loss')
plt.xlabel('epochs')
```

**Second experiment**: same as before, but 300 epochs

```python

tf.random.set_seed(42)

model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1)
])

model2.compile(loss = tf.keras.losses.mae,
                        optimizer = tf.keras.optimizers.Adam(), 
                        metrics = ["mae"])

history2 = model2.fit(X_train,y_train,epochs = 300, verbose = 0)
```

```python
ev2 = model2.evaluate(X_test,y_test)
```

```python
pd.DataFrame(history2.history).plot()
plt.ylabel('loss')
plt.xlabel('epochs')
```

```python
ev1,ev2
```

The result has improved.
